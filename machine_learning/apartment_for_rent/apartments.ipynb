When it comes to "pets_allowed" variable it seems that when this information is not provided then it means that animals are not allowed so we can substitute NaNs with negative response.


```python
# replacing NaN with "No" for pets_allowed variable:
df.loc[(df.loc[:, ('pets_allowed')].isna()), ('pets_allowed')] = "No"
```

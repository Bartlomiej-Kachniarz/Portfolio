Generating values from a function with `np.fromfunction`

```python
def real_function(a, b, c):
    return (a - b) ** c

def function_see_a(a, b, c):
    return a

def function_see_b(a, b, c):
    return b

def function_see_c(a, b, c):
    return c

print(np.fromfunction(real_function, (2, 3, 10)), "\n") # the shape = (2, 3, 10) and the a, b, c are following: 
```

```python
# How the a, b and c are represented:
print("a:\n", np.fromfunction(function_see_a, (2, 3, 10)), "\n") # a values
print("b:\n", np.fromfunction(function_see_b, (2, 3, 10)), "\n") # b values
print("c:\n", np.fromfunction(function_see_c, (2, 3, 10)), "\n") # c values
```

Singular Value Decomposition:

```python
ux
```

```python
U, S_diag, V = linalg.svd(ux)
```

```python
U.dot(linalg.inv(U)) # identity matrix
```

```python
S_diag # we need to convert this into full sigma matrix:
```

```python
S = np.zeros((3, 3))
S[np.diag_indices(3)] = S_diag
S
```

```python
V.dot(linalg.inv(V))
```

```python
ux - U.dot(S).dot(V) # = 0 
```

```python
np.trace(ux) # sum of diagonal elements
```

Function `solve` can solve set of equations:

* $7x + 2y = 16$
* $4x + 3y = -29$

```python
coefficients = np.array([[7, 2], [4, 3]])
values = np.array([16, -29])
solv = linalg.solve(coefficients, values)
solv
```

```python
coefficients.dot(solv) # checking the solution
```

```python
np.allclose(coefficients.dot(solv), values)
```
